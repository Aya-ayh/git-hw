## 1. Data Types in Python

Python has several built-in data types that are commonly used in data analysis.

### Numeric Types

- **int** – Integer numbers (e.g., `5`, `-42`)
- **float** – Decimal numbers (e.g., `3.14`, `-0.001`)
- **complex** – Complex numbers (rarely used in data analysis)



```python
a = 10         # int
b = 3.14       # float
```



### Boolean

- **bool** – Represents `True` or `False`



```python
is_valid = True
not_valid = False
print(is_valid , "," , type(is_valid))
print(not_valid , "," , type(not_valid))
```



### Text Type

- **str** – Strings or sequences of characters



```python
name = "Data Analysis"
print(type(name))
```

```python
help(print)   #you can use it to know everything about the function
```



### Sequence Types

- **list** – Ordered, mutable collection
- **tuple** – Ordered, immutable collection
- **range** – Represents a sequence of numbers



```python
numbers = [1, 2, 3]       # list
coordinates = (10, 20)    # tuple
```



### Set Types

- **set** – Unordered collection of unique elements



```python
fruits = {"apple", "banana", "apple"}  # Output: {'apple', 'banana'}
#unordered so there is no index
help(set)
#remember the set operations from discrete
# as uninion and intersaction(you can find them at help)
```



### Dictionary

- **dict** – Key-value pairs



```python
person = {"name": "Alice", "age": 25 ,a :b}
#such that a and b are variables we inform before
print(person["name"])
```

## 2. Loops in Python

Loops allow you to run blocks of code multiple times.

### For Loop

Used to iterate over a sequence (like a list or a string).



```python
print(range(10))
print(list(range(10)))
#range(start,end,step)
```

```python
for i in range(5):
    print("Iteration:", i)
```



### While Loop

Repeats as long as a condition is `True`.



```python
x = 0
while x < 3:
    print("x is", x)
    x += 1
```



### Loop Control Statements

- `break` – Exit the loop early
- `continue` – Skip the rest of the loop and go to the next iteration



```python
for i in range(5):
    if i == 3:
        break
    print(i)
```

## 3. Functions in Python

Functions allow you to reuse blocks of code.

### Defining a Function



```python
def greet(name):
    print("Hello,", name)
```



### Calling a Function



```python
greet("Alice")  # Output: Hello, Alice
```



### Return Statement



```python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)  # Output: 8
```



### Default Arguments



```python
def greet(name="friend"):
    print("Hi,", name)

greet()        # Output: Hi, friend
greet("Bob")   # Output: Hi, Bob
```



### Variable Scope

- Variables inside functions are **local**.
- Variables outside are **global** (unless shadowed).



```python
x = 5  # global

def change():
    x = 10  # local
    print("Inside function:", x)

change()
print("Outside function:", x)
```
