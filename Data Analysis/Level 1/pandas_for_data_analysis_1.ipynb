# 📊 **Introduction to Pandas**

Pandas is a powerful Python library used for working with data in an easy and flexible way. It helps you analyze, clean, and manipulate data quickly, making it one of the most important tools in Data Science.

Think of Pandas as Excel inside Python—but much more powerful and automated.

# **🚀 Why do we use Pandas?**

Pandas makes it easy to:

Load data from files (CSV, Excel, JSON, etc.)

Clean messy data (missing values, duplicates, errors)

Analyze and explore data

Transform and reshape data

Perform statistical calculations

Work with time-based data (dates and trends)

# **🧱 Main Data Structures**

Pandas is built on two main structures:

## **1. Series 📌**

A Series is like a single column of data.

## **2. DataFrame 📊**

A DataFrame is like a table (rows and columns), similar to Excel.

# **🔑 Key Features of Pandas**
Handles missing data easily

Allows fast filtering and searching

Supports grouping and aggregation

Can merge and join datasets like databases

Works smoothly with large datasets

```python
import pandas as pd
```

```python
help(pd)
```

## Step 1: Load Sample Data
We will create a small sample sales dataset manually to keep things simple.

```python

data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2026-05-04'],
    'Product': ['Apples', 'Bananas', 'Apples', 'Bananas', 'Orange'],
    'Quantity': [10, 15, 20, 5, 25],
    'Price': [0.5, 0.3, 0.5, 0.3, 0.7]
}
#the keys are the colunms and the values are the rows
df = pd.DataFrame(data)
df
#it will become as a list so there is index
```

## Step 2: Understand the Data

```python
help(df.info)
```

```python
df.info()
```

## Step 3: Descriptive Statistics
Get a quick summary of numeric columns.

```python
help(df.describe)
```

```python
df.describe()
```

## Step 4: Create a New Column
We can calculate `Revenue = Quantity × Price`.

```python
df['Revenue'] = df['Quantity'] * df['Price']
df
#or
print(df)
print(df.head(2))#to tell how many rows you want
print(df.tail(2))
# try  df['new column']=1
```

```python
help(df.head())
help(df.tail())
```

## Step 5: Filter Rows
Let's filter rows where Revenue is greater than 5.

```python
df[df['Revenue'] > 5]
```

```python
df[df['Product'] == 'Apples']
```

## Step 6: Group By Product
Total revenue per product.

```python
df.groupby('Product')['Revenue'].sum()
```

```python
df.groupby('Product')['Quantity'].sum()
```

## Step 7: Sorting the DataFrame
Sort by revenue in descending order.

```python
help(df.sort_values)
```

```python
df.sort_values(by='Revenue', ascending=False)
#from greater to less(desascending)
```

## Step 8: Handling Missing Values
Let's simulate missing data and handle it.

```python

df.loc[2, 'Quantity'] = None
print("With missing value:")
print(df)

# Fill missing value with 0
df['Quantity'] = df['Quantity'].fillna(0)
#na= not avilable
#ex.   df['Quantity'].fillna(df['Quantity'].medin()) or mode and so on
print("After filling missing value:")
print(df)
```
